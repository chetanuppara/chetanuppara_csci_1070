```python
import numpy as np
import pandas as pd 
```

```python
titanic_listings = ('titanic.csv')
titanic_df = pd.read_csv(titanic_listings)
titanic_df.head(50)
```

```python
### 1.When to Use ML .

#When You Have Lots of Data: ML works the best when there is a lot of data it can learn from. Suppose you have thousands of images of cats and dogs. An ML model can learn how to tell them apart. The more data you have, the better it can get.
#When Patterns Are Complex:The area in which ML excels most is when patterns are hidden within data and difficult for a human to easily deduce. Stock market prediction and speech recognition are examples of this, as they need lots of variables tested at once, which ML can do.
#Most importantly, it proves useful in situations where data is updated very frequently and things around us keep on changing. Because then, the model keeps updating or changing its output for new information. Recommendation systems, for instance, Netflix and Amazon make recommendations on what you have recently viewed or bought. Times You Want Predictions or Automation
#ML is good to do predictions, be those spam versus not spam emails, or automation of a task, say, face recognition in photos. If your goal is to have a system make decisions on its own based on data, ML could be the right tool.

#when not to use MLWhen Not to Use ML
#When Data is Scarce: ML requires data to learn patterns from. When you do not have enough data, the model will not be able to learn anything proper and hence may not give good results.
#When the Problem is Simple: If you can solve the problem through basic rules or logic, you have no need for ML. For instance, if you wanted to alphabetize some names, it would not be a good application of ML, since that would complicate something that was such an easily reached solution.
#When Accuracy Is Critical: Sometimes, ML models are simply wrong, and often they make mistakes. If your task really needs to be 100% accurate, such as medical diagnosis or safety-critical systems, probably the better thing to do would be to use something more reliable and tested.
#When You Can Use Traditional Algorithms:Sometimes, the performance of simpler traditional algorithms is better and often much quicker. In other words, if some non-ML approach works, giving you the answers that you want, there's no need to use ML.
```

```python
# 2...
# counts all the null values 
titanic_df.isnull().sum()
```

```python
#2...
# describe function is where it gives out statistical information 
titanic_df.describe()
print(titanic_df.head())
```

```python
#2...
#this function drops us columns that have null values 
titanic_df.dropna()
print(titanic_df.head())
```

```python
#2... this is the lambda function where it gives you values of what u need when its either value 1 or value 2
titanic_df['Cabin'] = titanic_df['Cabin'].apply(lambda x: 1 if x == 1 else 0)
print(titanic_df.head())
```

```python
#3
titanic_df.drop('Name', axis = 1, inplace = True)
titanic_df.drop('Sex', axis = 1, inplace = True)
titanic_df.drop('Ticket', axis = 1, inplace = True)
titanic_df.drop('Cabin', axis = 1, inplace = True)
titanic_df.drop('Embarked', axis = 1, inplace = True)
titanic_df.drop('PassengerId', axis = 1, inplace = True)

titanic_df
```

```python
#4...
import seaborn as sns 
import matplotlib.pyplot as plt 

sns.heatmap(titanic_df.corr(), annot = True)
plt.show()
```

```python
###4...
#Since we are building a correlation matrix, the columns Name, Sex, Ticket, Cabin, Embarked, and PassengerID were removed 
#since they are not useful in the data set. You cannot enter string values in the correlation matrix. 
#Negative values indicate a poor correlation between the variables that meet at that particular value in the matrix, 
#therefore we can disregard them. Since the variables that meet at a given value will be the same, the 1 in the correlation matrix simply 
#indicates that the variables are correlated with one another. Furthermore, we can conclude that there is some association between the 
#variables that exhibit positive correlation and that further research on them may be warranted. An illustration of a positive association

plt.boxplot(titanic_df['Fare'])
plt.title('box plot and whisker plot')
plt.show()
```
