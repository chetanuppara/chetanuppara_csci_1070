```python
import numpy as np
import pandas as pd
from sklearn import tree
from sklearn.metrics import ConfusionMatrixDisplay, classification_report

diabetes_df = pd.read_csv("diabetes.csv")

diabetes_df.head()
```

```python
from sklearn.model_selection import train_test_split

X = diabetes_df.drop('Outcome',axis=1)
y = diabetes_df['Outcome']

#split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X,y,
                                                    test_size=0.3,
                                                    random_state=42)
```

```python
# decision tree classifier
# deep and complex trees tend to overfit, so we want to
# generalize and have reasonably simple trees 
# as much as possible
model =tree.DecisionTreeClassifier(max_depth=4, random_state=42, min_samples_leaf=1)
```

```python
model = model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
```

```python
ConfusionMatrixDisplay.from_predictions(y_test, y_pred)
```

```python
# recall 
# tp / (tp + fn)
48/(48+32)
```

```python
import matplotlib.pyplot as plt
tree.plot_tree(model, feature_names=list(X.columns),
               filled=True, fontsize=6, rounded=True)

plt.show()
plt.savefig("tree.svg")
```

```python
def list_names(names):
    if not names:
        return ""
    list_string = ""
    counter = 0  
    
    while counter < len(names):
        if counter == len(names) - 2:  
            list_string += names[counter] + " & "
        elif counter == len(names) - 1: 
            list_string += names[counter]
        else:
            list_string += names[counter] + ", "
        counter += 1  
    
    return list_string

names = ["trump", "shaq", "meme", "chetan"]

formatted_names = list_names(names)
print(formatted_names)  
```
