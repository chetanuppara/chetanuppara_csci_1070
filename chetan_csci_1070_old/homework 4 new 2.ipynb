```python
import numpy as np
import matplotlib.pyplot as plt

# Load the data into the studentsurveydf variable
studentsurveydf = pd.read_csv('CSCI_Fall_2024_survey.csv')

def plot_major(CSCI_df, title='Major', xlabel='Major', ylabel='Count', rotation=90):
    major_counts = CSCI_df['What is your major?'].value_counts()
    major_index = major_counts.index
    x = np.array(major_index)
    y = np.array(major_counts)
    plt.bar(x, y)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.xticks(rotation=rotation)
    plt.show()

def plot_num_pets(CSCI_df, title='Number of Pets', xlabel='Number of Pets', ylabel='Count', bins=10):
    num_pets = CSCI_df['How many pets and of which kind do you have?']
    x = np.array(range(1, bins+1))
    y = np.array([sum(num_pets == i) for i in range(1, bins+1)])
    plt.bar(x, y)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()

def plot_age_vs_num_pets(CSCI_df, title='Age vs Number of Pets', xlabel='Age', ylabel='Number of Pets'):
    age = CSCI_df['How old are you?']
    num_pets = pd.to_numeric(CSCI_df['How many pets and of which kind do you have?'], errors='coerce')
    plt.scatter(age, num_pets)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()

def plot_favorite_food(CSCI_df, title='Favorite Food', xlabel='Food', ylabel='Count'):
    favorite_food = CSCI_df['What is your favorite type of food?'].value_counts()
    labels = favorite_food.index
    x = np.array(labels)
    y = np.array(favorite_food)
    plt.bar(x, y)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()

def plot_country_birth(CSCI_df, title='Country of Birth', xlabel='Country', ylabel='Count'):
    country_birth = CSCI_df['What state/country were you born in?'].value_counts()
    labels = country_birth.index
    x = np.array(labels)
    y = np.array(country_birth)
    plt.bar(x, y)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()

plot_major(CSCI_df)
plot_num_pets(CSCI_df)
plot_age_vs_num_pets(CSCI_df)
plot_favorite_food(CSCI_df)
plot_country_birth(CSCI_df)
```
